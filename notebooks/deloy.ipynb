```python
import os
def first_step():
    """
    Bước đầu xác định đã tạo dữ liệu
    Nếu chưa -> chạy đoạn còn lại
    """
    df = pd.DataFrame()
    if os.path.exists(PROCESSED_DATA_DIR / 'clean_data.csv'):
        df = load_clean_data()
    else:
        df = load_retail_data()

        df = clean_data(df)
        df = feature_engineering(df)
    
    # --- Thời gian cắt làm bộ dữ liệu tương lai
    cut_off_date = "2011-09-30"
    # --- Hồi quy
    if not os.path.exists(PROCESSED_DATA_DIR / 'regression_data.csv'):
        create_regression_or_classification_data(df,cut_off_date)
    # --- Phân loại
    if not os.path.exists(PROCESSED_DATA_DIR / 'classification_data.csv'):
        create_regression_or_classification_data(df,cut_off_date,is_classify=True)
    # --- Phân cụm
    if not os.path.exists(PROCESSED_DATA_DIR / 'rfm_data.csv'):
        create_rfm_data(df)
    # --- Chuỗi thời gian
    if not os.path.exists(PROCESSED_DATA_DIR / 'timeseries_data.csv'):
        create_timeseries_data(df)
    
    print("Dữ liệu đã tạo đầy đủ!\n")

first_step()
```
